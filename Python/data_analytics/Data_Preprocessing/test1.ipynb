```python
import pandas as pd
import numpy as np

df = pd.read_csv("https://raw.githubusercontent.com/YoungjinBD/dataset/main/airquality.csv")
df
```

##### 1. 결측치가 가장 많은 변수를 찾아서 해당 결측치를 0으로 대치하고, 결측치를 제외한 평균과 0을 대치한 후의 평균과의 차이를 구하시오

```python
print(df.isnull().sum())
```

```python
Ozone_mean = df['Ozone'].mean()
df['Ozone'].fillna(0, inplace=True)
Ozone_mean2 = df['Ozone'].mean()
print(abs(Ozone_mean-Ozone_mean2))
```

##### 2. Wind 변수에 대해 Min-max 정규화를 수행한 후, 평균값과 z정규화를 수행한 후 평균 값의 차이를 구하시오.

```python
# Min-max 정규화
Min = np.min(df['Wind'])
Max = np.max(df['Wind'])
df['min_max'] = round((df["Wind"]-Min)/(Max-Min), 2)

# z 정규화
Mean = np.mean(df["Wind"])
Std = np.std(df["Wind"])
df["z"] = round((df['Wind']-Mean) / Std,2 )

print(round(abs(df['min_max'].mean() - df['z'].mean()),2))
```

#### 3. 월별(5월~9월) 평균 기온을 구하시오.

```python
df.groupby("Month").mean()['Temp'].round(2).to_csv('result.csv', index=False, encoding='utf-8')
```

##### 2.BMI는 몸무게(kg)를 키(m)의 제곱으로 나누어 계산한다.
- 주어진 BMI 데이터셋에서 비만도가 정상에 속하는 인원 수와 과체중에 속하는 인원 수의 차이를 정수로 출력
- 정상 : 18.5 이상, 23 미만
- 과체중 : 23이상 25미만

```python
df = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/dataset/main/BMI.csv')
df.head()
```

```python
df['BMI'] = df['Weight'] / ((df['Height']/100)**2)
normal = len(df[(df['BMI'] >= 18.5)&(df['BMI']<23)])
over_weight = len(df[(df['BMI'] >= 23)&(df['BMI']<25)])
print(int(abs(normal - over_weight)))
```

##### 3. 순 전입학생이 가장 많은 학교의 전체학생 수를 구하시오.

```python
df = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/dataset/main/students.csv', encoding='cp949')
df.head()
```

```python
df = df.groupby('학교').sum()
df['순 전입학생'] = df['총 전입학생'] - df['총 전출학생']
max_num = -1
idx = -1
for i in range(0,5):
    if df.iloc[i]['순 전입학생'] > max_num:
        idx = i
        max_num = df.iloc[i]['순 전입학생']

print(df.iloc[idx]['전체학생 수'])
```
